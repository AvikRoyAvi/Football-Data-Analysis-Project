```python
#Import the libraries
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#Load dataset
df = pd.read_csv('D:\Python codes\Football Data Analysis Project using python\EPL_20_21.csv')
df
```

```python
df.info
```

```python
df.describe()
```

```python
df.isna().sum()
```

```python
#Create 2 more columns
df['MinsPerMatch'] = (df['Mins'] / df['Matches']).astype(int)
df['GoalsPerMatch'] = (df['Goals'] / df['Matches']).astype(float)
df.head()
```

```python
df
```

```python
#Total Goals
total_goals = df['Goals'].sum()
print(total_goals)
```

```python
#Penalty Goals
penalty_goals = df['Penalty_Goals'].sum()
print(penalty_goals)
```

```python
#Penalty Attempted
penalty_attempted = df['Penalty_Attempted'].sum()
print(penalty_attempted)
```

```python
#Pie chart for penalties missed VS scored
plt.figure(figsize = (12, 6))
not_scored = df['Penalty_Attempted'].sum() - penalty_goals
data = [not_scored, penalty_goals]
labels = ['Penalties missed', 'Penalties scored']
colors = ['r', 'g']
textprops = {"fontsize": 15}
plt.pie(data, labels = labels, colors = colors, autopct = '%0.1f%%', shadow = True, radius = 2, textprops = textprops)
plt.legend()
plt.show()
```

```python
#Unique position
df['Position'].unique()
```

```python
#Total FW players
df[df['Position'] == 'FW']
```

```python
#Player from different nation
np.size(df['Nationality'].unique())
```

```python
#Most players from which country
nationality = df.groupby('Nationality').size().sort_values(ascending = False)
nationality.head(10).plot(kind = 'bar', figsize = (12, 6), color = sns.color_palette('hot'))
plt.show()
```

```python
#Club of maximum players in their squad
df['Club'].value_counts().nlargest(5).plot(kind = 'bar', color = sns.color_palette('rainbow'), figsize = (12, 6))
plt.show()
```

```python
#Club of least players in their squad
df['Club'].value_counts().nsmallest(5).plot(kind = 'bar', color = sns.color_palette('twilight'), figsize = (12, 6))
plt.show()
```

```python
#Players based on age group
under_20 = df[df['Age'] <= 20]
age_20_25 = df[(df['Age'] > 20) & (df['Age'] <= 25)]
age_25_30 = df[(df['Age'] > 25) & (df['Age'] <= 30)]
above_30 = df[df['Age'] > 30]
```

```python
x = np.array([under_20['Name'].count(), age_20_25['Name'].count(), age_25_30['Name'].count(), above_30['Name'].count()])
plt.figure(figsize = (12, 6))
labels = ['<=20', '>20&<=25', '>25&<=30', '>30']
textprops = {"fontsize": 12}
plt.title('Total players with age', fontsize = 20)
plt.pie(x, labels = labels, autopct = '%.1f%%', textprops = textprops)
plt.show()
```

```python
#Total under 20 players in each club
plt.figure(figsize = (16, 9))
players_under_20 = df[df['Age'] < 20]
players_under_20['Club'].value_counts().plot(kind = 'bar', color = sns.color_palette('ocean'))
plt.show()
```

```python
#Under 20 players in manU
players_under_20[players_under_20['Club'] == 'Manchester City']
```

```python
#Under 20 players in Chelsea
players_under_20[players_under_20['Club'] == 'Chelsea']
```

```python
#Average age of players in each group
plt.figure(figsize = (16,9))
sns.boxplot(x = 'Club', y = 'Age', data = df)
plt.xticks(rotation = 90)
plt.show()
```

```python
num_players = df.groupby('Club').size()
data = (df.groupby('Club')['Age'].sum()) / num_players
data.sort_values(ascending = False)
```

```python
#Total assists from each club
assist_by_club = pd.DataFrame(df.groupby('Club', as_index = False)['Assists'].sum())
sns.set_theme(style = 'whitegrid', color_codes = True)
ax = sns.barplot(x = 'Club', y = 'Assists', data = assist_by_club.sort_values(by = 'Assists'), palette = 'spring')
ax.set_xlabel('Club', fontsize = 30)
ax.set_ylabel('Assists', fontsize = 30)
plt.xticks(rotation = 75)
plt.rcParams['figure.figsize'] = (20, 10)
plt.title('Plot of Club VS Assists', fontsize = 20)
plt.show()
```

```python
#Top 10 assists
top_10_assists = df[['Name', 'Club', 'Assists', 'Matches']].nlargest(n = 10, columns = 'Assists')
top_10_assists
```

```python
#Goals by club
Golas_by_club = pd.DataFrame(df.groupby('Club', as_index = False)['Goals'].sum())
sns.set_theme(style = 'whitegrid', color_codes = True)
ax = sns.barplot(x = 'Club', y = 'Goals', data = Golas_by_club.sort_values(by = 'Goals'), palette = 'rocket')
ax.set_xlabel('Club', fontsize = 30)
ax.set_ylabel('Goals', fontsize = 30)
plt.xticks(rotation = 75)
plt.rcParams['figure.figsize'] = (20, 10)
plt.title('Plot of Club VS Total Goals', fontsize = 20)
plt.show()
```

```python
#Most goals by players
top_10_goals = df[['Name', 'Club', 'Goals', 'Matches']].nlargest(n = 10, columns = 'Goals')
top_10_goals
```

```python
#Goals per match
top_10_goalspermatch = df[['Name', 'GoalsPerMatch', 'Matches', 'Goals']].nlargest(n = 10, columns = 'GoalsPerMatch')
top_10_goalspermatch
```

```python
#Pie chart - goals with assists and without assists
plt.figure(figsize = (16, 9))
assists = df['Assists'].sum()
data = [total_goals - assists, assists]
textprops = {"fontsize": 12}
labels = ['Goals without assists', 'Goals with assists']
colors = sns.color_palette('Set1')
plt.pie(data, labels = labels, colors = colors, autopct = '%.0f%%', textprops = textprops)
plt.show()
```

```python
#Top 10 players with most yellow cards
yellow_card = df.sort_values(by = 'Yellow_Cards', ascending = False)[:10]
plt.figure(figsize = (20, 10))
plt.title('Players with most yellow cards', fontsize = 20)
yc = sns.barplot(x = yellow_card['Name'], y = yellow_card['Yellow_Cards'], label = 'Players', color = 'yellow')
plt.ylabel = ('Number of yellow cards')
yc.set_xticklabels(yc.get_xticklabels(), rotation = 45)
yc
plt.show()
```
